# Entendimento do Negócio (Business Understanding)

## 1. Contexto do Negócio

Empresas de varejo e e-commerce enfrentam desafios constantes na gestão de estoque e tomada de decisão de compras. A ausência de modelos preditivos leva a decisões baseadas apenas em histórico simples ou intuição, resultando em:

- Excesso de estoque (capital parado)
- Ruptura de produtos (perda de vendas)
- Aumento de custos logísticos e operacionais

Com o crescimento do volume de dados, torna-se possível aplicar técnicas de Data Science para transformar dados históricos em decisões estratégicas.

Este projeto utiliza dados do dataset do site Kaggle:
[Brazilian E-Commerce Public Dataset by Olist](https://www.kaggle.com/datasets/terencicp/e-commerce-dataset-by-olist-as-an-sqlite-database)

com o objetivo de desenvolver um sistema inteligente de análise financeira e consumo inteligente.

## 2. Problema de Negócio

Como prever a demanda futura de produtos e determinar o momento e a quantidade ideal de comprar, minimizando custos e evitando ruptura de estoque?

## 3. Objetivo do Projeto

Desenvolver um sistema inteligente capaz de:

- Prever a demanda futura de produtos
- Identificar o momento ideal para reposição de estoque
- Recomendar a quantidade ideal de compra
- Gerar insights financeiros para otimização de custos

O foco principal é aplicar o conceito de **consumo inteligente**, otimizando o uso de recursos financeiros e operacionais, recomendando a quantidade ideall, reduzindo desperdícios e maximizando eficiencia financeira.

## 4. Questões de Negócio

- Qual será a demanda futura de um determinado produto?
- Quando devemos realizar uma nova compra?
- Qual a quantidade ideal a ser comprada?
- Quais produtos apresentam maior risco de ruptura?
- Quais produtos geram maior custo de estoque?

## 5. Hipóteses

**H1:** Produtos com alta variabilidade de vendas necessitam maior estoque de segurança

**H2:** A previsão de demanda reduz significativamente o risco de ruptura

**H3:** Compras baseadas em previsão reduzem o custo total de estoque

**H4:** Existe sazonalidade nas vendas de determinados produtos

## 6. Consumo Inteligente

Neste projeto, consumo inteligente é definido como:

> A utilização eficiente de recursos financeiros e operacionais, baseada em dados, para otimizar decisões de compra e gestão de estoque.

Isso inclui:
- Redução de desperdício
- Otimização de capital investido
- Melhoria na disponibilidade de produtos

## 7. Métricas

### 7.1 Modelo
- MAE (Mean Absolute Error)
- RMSE (Root Mean Squared Error)

### 7.2 Negócio
- Redução de ruptura (%)
- Redução de estoque parado (%)
- Giro de estoque
- Custo total estimado

### 7.3 Financeiro
- Economia potencial gerada

### 7.4 Ecológicas
- Redução de desperdício (%)
  → Produtos não vendidos ou descartados 

- Estoque excedente médio
  → Volume de produtos armazenados sem necessidade

- Emissões estimadas de CO₂ (simulado)
  → Baseado em transporte e armazenagem

- Eficiência de abastecimento
  → Relação entre produtos vendidos e comprados

## 8. Estratégia de Solução

O projeto será dividido em etapas:

1. Coleta e entendimento dos dados
2. Preparação e tratamento
3. Feature Engineering
4. Modelagem preditiva (séries temporais)
5. Simulação de estoque
6. Geração de insights e recomendações

## 9. Pipeline do Projeto

Dados → Processamento → Features → Modelo → Previsão → Regras de Negócio → Insight

## 10. Premissas

Devido à ausência de algumas variáveis no dataset, serão consideradas:

- Lead time fixo (simulado)
- Estoque inicial estimado
- Custo de armazenagem aproximado
- Demanda baseada em histórico de vendas

Essas premissas permitem a construção de um modelo aplicável mesmo com limitações de dados.

## 11. Limitações

- Dataset não possui dados reais de estoque
- Não há informações diretas de fornecedores
- Custos logísticos são estimados
- Possível ausência de sazonalidade completa

Essas limitações serão tratadas através de simulações e suposições controladas.

## 12. Valor Gerado

Este projeto gera valor ao integrar análise preditiva e regras de negócio para otimização do consumo de recursos, proporcionando:

- Redução de custos operacionais, por meio da diminuição de excesso de estoque e compras ineficientes
- Aumento da eficiência na cadeia de suprimentos, com melhor planejamento de reposição
- Melhoria na tomada de decisão, baseada em dados e previsões de demanda
- Maior disponibilidade de produtos, reduzindo riscos de ruptura
- Otimização do capital de giro, evitando alocação desnecessária de recursos em estoque
- Redução de impactos ambientais, através da diminuição de desperdícios e melhor gestão de recursos
- Fortalecimento da imagem da empresa, alinhada a práticas de consumo inteligente e sustentabilidade
